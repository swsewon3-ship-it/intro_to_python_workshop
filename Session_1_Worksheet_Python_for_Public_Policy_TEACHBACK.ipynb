### Option 1: Population Growth (Core)

````
population_2020 = 50_000_000
population_2025 = 52_000_000

growth = (population_2025 - population_2020) / population_2020 * 100

if growth < 5:
    label = 'Low growth'
elif growth < 15:
    label = 'Medium growth'
else:
    label = 'High growth'

print(label + ':', round(growth, 2), '%')
````
